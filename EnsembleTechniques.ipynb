```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

DATA = {
    "csv": ["./TelcomCustomer-Churn_1.csv", "./TelcomCustomer-Churn_2.csv"],
    "csv_merge_on": "customerID",
    "identity_columns": ["customerID"],
    "pie_chart_columns": ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 
       'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
       'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
       'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod'],
    "categorical_data": {
        "Churn":{"No":0, "Yes":1}
    },
    "target_column": "Churn"
}

def get_dataframe_from_csv(data):
    df = pd.DataFrame()
    if len(data["csv"]):
        for item in data["csv"]:
            csv = pd.read_csv(item)
            if len(df):
                df = df.merge(csv, on = data["csv_merge_on"], how = 'left')
            else:
                df = pd.read_csv(item)
    else:
        df = pd.read_csv(item)

    if df.empty:
        print('DataFrame is empty!')
    else:
        print("CSV shape for ", data["csv"], " : ", df.shape)

    return df


def process_null_and_identity_columns(df, identity_columns):
    # dropping identity columns which doesnt contribute to data much
    df.drop(columns=identity_columns, inplace=True)
    for item in df.columns:
        if df[item].dtypes == "float64" or df[item].dtypes == "int64":
            df[item] = df[item].apply(pd.to_numeric, errors = 'coerce')

    # dropping na values
    df.dropna()
    return df

def drawPieChart(df, columns):
  plt.figure(figsize=(25,25))
  for index, col in enumerate(columns):
    plt.subplot(7, 3, index+1)
    df[col].value_counts().plot(kind='pie', autopct='%1.0f%%')

def data_encoding(df, categorical_data):
    cat_data = categorical_data.keys()
    for item in cat_data:
        if item in df.columns:
            df = df.replace(categorical_data[item])

    colsToEncode=[]
    for feature in df.columns:
        if df[feature].dtype=='object':
            colsToEncode.append(feature)
        elif df[feature].dtype!='int64' and df[feature].dtype!='float64':
            colsToEncode.append(feature)

    #getting dummies for the remaining columns 
    df = pd.get_dummies(df, columns=colsToEncode)
    return df

def standardize_data(X_train, X_test):
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    return X_train_scaled, X_test_scaled

def create_evalset_data(df, target_col):
    X = df.drop(columns=target_col, axis=1)
    y = df.pop(target_col)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=7)
    X_train_scaled, X_test_scaled = standardize_data(X_train, X_test)
    evalset = [(X_train_scaled, y_train), (X_test_scaled, y_test)]
    return X_train_scaled, X_test_scaled, y_train, y_test, evalset

# Data pre processing returns -> X_train_scaled, X_test_scaled, y_train, y_test, evalset
def preprocessing_(data):
    # fetch data from CSV as a dataframe
    df = get_dataframe_from_csv(data)
    # Processing null values
    df = process_null_and_identity_columns(df, data["identity_columns"])
    # Pie chart for percentage calculations
    drawPieChart(df, data["pie_chart_columns"])
    # Data encoding with categorical values
    df = data_encoding(df, data["categorical_data"])
    # Generating eval_set with train test split
    evalset = create_evalset_data(df, data["target_column"])
    return evalset

def create_model(X_train_scaled, y_train, evalset, n_estimators=15):
    model = XGBClassifier(n_estimators=n_estimators, eta=0.05)
    model.fit(X_train_scaled, y_train, eval_metric=["error", "logloss"], eval_set=evalset)
    return model

# Prints Accuracy, plot learning curve and return predictions
def make_predictions(model, X_test_scaled, y_test):
    y_pred = model.predict(X_test_scaled)
    predictions = [round(value) for value in y_pred]
    score = accuracy_score(y_test, y_pred)
    return score, predictions

def main():
    X_train_scaled, X_test_scaled, y_train, y_test, evalset = preprocessing_(DATA)
    model = create_model(X_train_scaled, y_train, evalset)
    score, predictions = make_predictions(model, X_test_scaled, y_test)
    print("Accuracy of model is ", score)

main()

```
